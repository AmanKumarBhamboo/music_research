```python

import librosa
import numpy as np
import matplotlib.pyplot as plt
import sounddevice as sd
```

```python
record_sr = 44100
process_sr = 11025
duration = 5
```

```python
print("Recording...")
audio = sd.rec(int(duration * record_sr), samplerate=record_sr, channels=1)
sd.wait()
print("Recording finished")

audio = audio.flatten()
```

```python
time = np.linspace(0, len(audio)/record_sr, num=len(audio))
```

```python
import librosa.display
plt.figure(figsize=(12,4))
librosa.display.waveshow(audio, sr=record_sr)

plt.title("Audio Waveform")
plt.show()
```
